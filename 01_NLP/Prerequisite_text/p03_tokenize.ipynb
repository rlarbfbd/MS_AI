```python
import nltk                                 # 개별 기능은 따로 설치

nltk.download("punkt_tab")                  #  특수문자 관련
```

```python
sentence = "우리 가비아 과정은 AI 애플리케이션 개발자 직군이고 직무는 어디로 빠질까요? AI 어플리케이션(APP) 개발자?"

from nltk.tokenize import word_tokenize, sent_tokenize

a = word_tokenize(sentence)                 # 단어 나누기 (특수문자 분리)
print(a)

b = sent_tokenize(sentence)                 # 문장 나누기
print(b)
```
