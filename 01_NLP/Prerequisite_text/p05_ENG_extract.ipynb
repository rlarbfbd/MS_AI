```python
# 어간 추출(동사원형 X)

from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()                    # 어간 추출(동사원형X)

words = "play playing played plays"
for w in word_tokenize(words):
    print(w)
    print(ps.stem(w))
    print("--------------------")

words2 = "variable variety variation"
for w2 in word_tokenize(words2):
    print(w2)
    print(ps.stem(w2))
    print("--------------------")
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import wordnet

wnl = WordNetLemmatizer()                   # 동사원형

words = "am are is"
for w in word_tokenize(words):
    print(w)
    print(wnl.lemmatize(w, wordnet.VERB))
    print("--------------------")

words2 = "say says said"
for w2 in word_tokenize(words2):
    print(w2)
    print(wnl.lemmatize(w2, wordnet.VERB))
    print("--------------------")
```
