```python
c = "설"

from jamo.jamo import h2j, j2hcj

d= h2j(c)                       # 조합형 자음/모음으로 분리
print(d)                        # VScode에서 제대로 안 보임

e = j2hcj(d)                    # 조합형 자음/모음 -> 온전한 자음/모음
print(e)
print(e.find("ㅓ"))
```
