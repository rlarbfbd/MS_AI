```python
# 무슨 단어가 몇 번 나오냐 -> Word Count/ BoW(Bag of Words) : 텍스트 처리 기본
wc = {}
with open("./Users/stu_ict01_01/sam/sam01.txt", "r", encoding="utf=8") as f:
    for line in f.readlines():
        line = line.replace("\n", "")
        line = line.split(" ")

        for word in line:
            if word in wc:
                wc[word] += 1
            else:
                wc[word] = 1
print(wc)                              # 여기서는 아무 의미없는 띄어쓰기까지 셈 & 명사와 조사가 합쳐진 채로 카운트
```
