```python
a = ["ㄱ", "ㄴ", "ㄷ"]
b = ["ㄱ", "ㄷ", "ㄹ", "ㅂ"]
```

```python
# b에서 a에도 있는 것만 출력
for bb in b:            # in: for문 문법
    if bb in a:         # in: list에 포함되어 있는지; 논리연산자
        print(bb)
```

```python
# b에 있는 거를 c에도 똑같이
# c = []
# for bb in b:
#     c.append(bb)

c = [bb for bb in b]
print(c)
```

```python
# b에 있는 거 중에 a에도 있는 것만 d에
# d = []
# for bb in b:
#     if bb in a:
#         d.append(bb)

d = [bb for bb in b if bb in a]
print(d)
```
