```python
sentences = "클라우드 환경 위에 ai 플랫폼 구성해서 사업하고 있는 듯 해요"
sentences += "선생님 봤을 때 이거 난이도가 어때요?"
sentences += "비전공자 기준과 우리 개발자 과정 수료생 듣는다 했을때?"
sentences += "에잌ㅋㅋㅋㅋㅋ"

from konlpy.tag import Okt

o = Okt()                               # Open Korean Text 분석기 (구 트위터한글형태소분석기)

a = o.normalize(sentences)              # 정규화(정리)
print(a)

b = o.phrases(sentences)                # 어구 추출(의미있는 단위로 나눔)
print(b)

c = o.morphs(sentences)                 # 형태소 분석(합성어도 나눔)
print(c)

d = o.morphs(sentences, stem=True)      # 형태소 분석(합성어도 나눔 + 동사원형)
print(d)

print("\n")

#e = o.pos(sentences)                   # 형태소 분석(합성어도 나눔) -> (단어, 품사) 리스트
e = o.pos(sentences, stem=True)
for ee, pp in e:
    print(ee)
    print(pp)
    print("-----")
```
