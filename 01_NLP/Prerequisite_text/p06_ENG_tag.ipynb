```python
# 품사 태깅
import nltk
nltk.download("averaged_perceptron_tagger_eng")
```

```python
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
# pos_tag: 단어 리스트를 넣으면 (단어, 품사) 투플로 만들어줌

sentences = "Then it was not Arthur who took them?"
wt = word_tokenize(sentences)

pt = pos_tag(wt)

for w, p in pt:
    print(w)
    print(p)
    print("------------")
```
