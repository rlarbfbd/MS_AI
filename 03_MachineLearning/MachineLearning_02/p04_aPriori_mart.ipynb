```python
# a priori 알고리즘
#   비지도 학습
#   A가 나왔을 때 B가 나올 확률을 구해주는 솔루션? 수학식?
#   -> 연관 관계 높다 -> 추천 시스템
#   머신러닝 알고리즘이 맞나? -> scikit-learn에 미포함

# supportData(지지도)
#   0.3으로 세팅 -> 100명 중 30명 이상이 구매한 물품만 계산하기 시작하자
#   a 나오고 b나올 확률 x
#   c 나오고 d 나올 확율 y
#   ...
#   지지도 못 넘는거 빼고
#   두 개씩 붙여서 또 계산하고, 지지도 못 넘는거 빼고 반복

# confienceData(신뢰도)
#   0.7로 세팅 -> 계산 결과가 0.7 넘는 것만 최종 결과로 보여줌
```

```python
feature = [
    ["치킨", "치킨", "피자", "소주"],
    ["맥주", "피자", "소주"],
    ["치킨", "맥주", "피자", "소주"],
    ["맥주", "막걸리"]
]
```

```python
from apyori import apriori

result = apriori(feature, min_support=0.3, min_confidence=0.5)
result = list(result)
for r in result:
    for r2 in r.ordered_statistics:
        print(r2.items_base, "를 산 사람이 ")
        print(r2.items_add, "를 살 확률이 ")
        print(r2.confidence)
        print(r2.lift)                          # apyori 라이브러리에만 있는 것 a priori 알고리즘에는 없음
                                                # 치킨 -> 맥주 살 확률/ 각각 따로 살 확률
                                                # 따로 나올 확률보다 같이 나올 확률이 얼마나 되는가
        print("-----------")
```
