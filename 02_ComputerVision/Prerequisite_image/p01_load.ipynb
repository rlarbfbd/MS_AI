```python
import cv2

# GRAYSCALE: 흑백 픽셀을 표현(흰색: 255, 검정: 0)
a = cv2.imread("./Users/stu_ict01_01/May28_2_Prerequisite_image/134093712144418504.jpg", cv2.IMREAD_GRAYSCALE)
print(a)

# COLOR: 컬러 픽셀 표현(픽셀 하나가 리스트가 되어 그 안에 BGR 순서대로 값)
b = cv2.imread("./Users/stu_ict01_01/May28_2_Prerequisite_image/134093712144418504.jpg", cv2.IMREAD_COLOR)
b = cv2.cvtColor(b, cv2.COLOR_BGR2RGB)      # BGR -> RGB
print(b)
```
