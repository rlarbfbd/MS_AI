```python
# 최적의 a, b를 찾자 -> 쌩랜덤으로 할 순 없음
# -> Gradient Descent(경사하강법)
# Gradient는 항상 값이 가장 커지는 방향을 향하니
# 반대로 가면 오차(MSE)를 줄이는 쪽으로 감
# L = e1 + e2 + e3 + ... + e10000
#  - 모든 데이터를 다 고려 -> 너무 느림
#  - local minimun

# SGD(Stochastic GD)
#   랜덤하게 하나만 해서 e3(모든 데이터 다 X)
#   남은 것 중에 또 랜덤한 하나만 해서 e4
#   ...
#   -> 하나만 하니 -> 빠름, 의외성이 생겨서 local minimum을 벗어날 수도 있음
#   -> 단점:
#       의외성이 너무 심함(답과는 전혀 다른 것이 나올 수 있음), 너무 성급

# momentum: GD갈 때 그 전 단계의 G값을 추가로 계산에 넣어서 : 관성
# -> 움직이는 값이 줄음( 더 성급하지 않게 답을 향해감)

# RMSProp(RootMeanSquared Propagation)
#   x, y, ... 각각 누적된 G 값으로 나눠서
#   -> x, y, ...를 동등하게 만들어서 의외성을 노려보자

# Adam(ADAptive Momentum estimation)
#   GD 기반 => Momentum + RMSProp
#   움직이는 방향은 momentum
#   움직이는 값은 RMSProp으로 평탄하게


# mini-batch SGD
#   랜덤하게 몇 개씩
#   batch_size: 한 번에 데이터 몇 개씩 볼거냐

# parameter: AI가 스스로 찾아내야할 값(Weight, bias)
# hyperparameter: 사람이 지정해줘야 할 값(learning_rate: GD갈 때 얼마씩 갈거냐, batch_size, epoch: 전체 데이터를 다 보는 그 작업 자체를 몇 번 반복할 것이냐, AI 모델 구조)
```
