Linear regression: 주어진 데이터들을 최대한 만족 시킬만한 a, b를 찾아서 회귀방정식(y = ax + b)을 완성시키고 그 식을 써서 예측

인공신경망(DNN): (다차원 행렬계산 + 다층) 다층 회귀분석 

y = ax + b
[1, 10], [2, 20], [3, 30] -> [4, ?] => y = 10x

[1, 9], [2, 19.3], [3, 31] -> [4, ?]

```python
import tensorflow as tf

tf.compat.v1.disable_v2_behavior()
```

```python
# 1) 훈련 데이터 확보 + 처리
xData = [1, 2, 3]
yData = [10, 20, 30]
```

```python
# 2) 인공신경망 구성(y = ax + b)
# 변수 종류
#   Variable: AI가 찾아내야 할 값(a, b)
#   placeholder: 훈련 데이터 들어갈 자리(x, y)

# 첫 값(0, 1차원, 자료형은 64bit float)
# 64bit float가 소수점이하 8자리 표현된다 치면
# 소수점 이하 8자리까지 필요한 문제인가?
# 그 정도까지 필요하지 않으면 -> 8bit float 사용하면 메모리 사용량 줄음
# => AI모델 양자화
a = tf.Variable(tf.zeros([1], dtype=tf.float64))         # 첫 시작을 y = 0.0000000x
# 첫 값(-1 ~ 2 랜덤(정규분포))
b = tf.Variable(tf.compat.v1.)

x = tf.compat.v1.placeholder(tf.float64)                 # 옛날 버전꺼라 compat.v1 작성
y = tf.compat.v1.placeholder(tf.float64)

sik = x * a + b
# sik을 통해 구한 값과 실제 값 차이가 크다면 a, b 바꾸서 다시 실행 -> 반복

# MSE + 차원 수 줄임
loss = tf.reduce_mean(tf.square(y - sik))

# 최적의 a, b를 찾아줄 객체
# learning+rate가 너무 크면 : 빠르지만 답을 못 찾고 지나가버릴  수도 있음
# learning+rate가 너무 작으면 : 답은 찾겠지만 느림
# -> 실전에서는 계산할 수 있는 수준은 아닐 것임 => 학습 돌려서 분위기 보고 바꿔가면서 찾아야함
o = tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.1)
#o = tf.compat.v1.train.AdamOptimizer(learning_rate=0.1)            # 실제로 가장 많이 씀

# loss를 최소화하는게 목표
goal = o.minimize(loss)
```

```python
# 3) 훈련 데이터를 넣어서 최적의 a, b 찾기: 학습/훈련
s = tf.compat.v1.Session()
s.run(tf.compat.v1.global_variables_initializer())      # variable들 첫 값 세팅

for i in range(1000):                                                  # epoch: 전체 데이터를 다 보는 그 작업 자체를 몇 번 반복할 것이냐
    s.run(goal, feed_dict={x:xData, y:yData})                          # placeholder x에 xData, y에 yData
    #print(s.run(loss, feed_dict={x:xData, y:yData}))                  # 이 값이 줄어가면 잘 학습되고 있음
    print(f"y = {s.run(a)[0]}x + {s.run(b)[0]}")                       # 지금 하고있는 것은 내가 값을 알고있으니 잘 되가는지 직접 식으로 확인하기
```

```python
# 4) 완성된 알고리즘을 활용해 미지의 데이터에 대한 결과 예측
newX = float(input("x: "))
newX = [newX]
result = s.run(sik, feed_dict={x:newX})
print(result[0])
```
