```python
import tensorflow as tf

tf.compat.v1.disable_v2_behavior()      # TF 2.x кё°лҠҘ OFF
```

```python
# мқёкіөмӢ кІҪл§қ л§Ңл“Өм–ҙм„ң нӣҲл Ё, мҳҲмёЎ мһ‘м—…м—җл§Ң нҠ№нҷ”
# 1) мқёкіөмӢ кІҪл§қ кө¬м„ұ                     # -> мӢӨм „м—җм„ңлҠ” 1лӢЁкі„к°Җ нӣҲл ЁлҚ°мқҙн„° нҷ•ліҙ + м „мІҳлҰ¬
a = tf.constant(10)                     # ліҖмҲҳ X, мғҒмҲҳ
b = tf.constant(5)
c = tf.add(a, b)                        # a + bлҘј н•ҳкІ лӢӨкі  мӢқл§Ң л§Ңл“Өм–ҙл‘  (кі„мӮ°X)
print(a, b, c)

d = tf.constant("г…Ӣ")
print(a, b, c, d)
```

```python
# 2) нӣҲл Ё лҚ°мқҙн„°лҘј л„Јм–ҙм„ң мөңм Ғмқҳ м•Ңкі лҰ¬мҰҳ м°ҫкё°
s = tf.compat.v1.Session()              # мӢӨн–ү м„ём…ҳ
# aResult = s.run(a)                      # мӢӨн–ү м„ём…ҳм—җ л„Јкі  лҸҢлҰ¬л©ҙ мӢӨн–үлҗҳм–ҙм„ң к°’ л¶Җм—¬лҗЁ
# print(aResult)

# dResult = s.run(d)
# print(dResult.decode())                 # н•ңкёҖ лҚ°мқҙн„°лҠ” decode н•„мҡ”

cResult = s.run(c)                      # м—°кҙҖлҗҳм–ҙ мһҲлҠ” кІғ мһҗлҸҷмңјлЎң м „л¶Җ мӢӨн–үн•ҙмӨҢ
print(cResult)

# 3) мҷ„м„ұлҗң м•Ңкі лҰ¬мҰҳмқ„ нҷңмҡ©н•ҙ лҜём§Җмқҳ лҚ°мқҙн„°м—җ лҢҖн•ң кІ°кіј мҳҲмёЎ
```
