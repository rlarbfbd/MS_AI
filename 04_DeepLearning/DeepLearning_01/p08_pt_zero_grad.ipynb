```python
import torch
```

```python
a = torch.tensor(5.0, requires_grad=True)           #5에서 시작해서 GD써서 답을 찾아가기
o = torch.optim.SGD([a], lr=0.1)
```

```python
for epoch in range(10):
    sik = 2 * a

    o.zero_grad()           # 누적 되어있을 전 단계의 G값을 0으로
    sik.backward()
    print(a.grad.item())        
    # sik을 a에 관해 편미분: 2
    # 편미분한거를 벡터로 쌓은 것을 Gradient: [2]
    # 2가 나와야하지만 출력 값을 보면 2, 4, 6, ...
    # why? => PyTorch는 전 단계의 G값을 누적시킴 => 우리는 이 기능이 필요없음 => 그래서 zero_grad() 사용
```
