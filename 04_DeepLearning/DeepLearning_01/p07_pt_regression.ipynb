```python
# TensorFlow 1.x: мҷ„м „ мҲҳмһ‘м—…

# PyTorch: м–ҙлҠҗ м •лҸ„ м•Ңм•„м•јн•ҳкі  м–ҙлҠҗ м •лҸ„ м»ЁнҠёлЎӨ н•ҙмӨҳм•јн•Ё

# TensorFlow 2.x: мҷ„м „ мһҗлҸҷ

# TensorFlow: мғқмӮ°л Ҙ -> нҳ„м—…
# PyTorch: м»ЁнҠёлЎӨ -> м—°кө¬
# => к·ёлһҳлҸ„ мҡ”мҰҳмқҖ PyTorchк°Җ лҢҖм„ё
%pip install torch
```

```python
import torch
```

```python
xData = torch.FloatTensor([[1], [2], [3]])
yData = torch.FloatTensor([[12], [22], [32]])
```

```python
a = torch.zeros(1, requires_grad==True)         # 1м°Ёмӣҗ, GD нҶөн•ҙм„ң м°ҫм•„лӮҙм•јн•  к°’мқҙлқјкі  лӘ…мӢң
b = torch.zeros(1, requires_grad==True)         # 1м°Ёмӣҗ, GD нҶөн•ҙм„ң м°ҫм•„лӮҙм•јн•  к°’мқҙлқјкі  лӘ…мӢң

o = torch.optim.SGD([a, b], lr=0.1)             # SGDмҚЁм„ң a, b к°’ м°ҫм•„лӮҙлқј, learning_rate=0.1
```

```python
for epoch in range(1000):
    sik = xData * a + b
    l = torch.mean((sik - yData) ** 2)      #MSE

    o.zero_grad()   # ??
    l.backward()    # м—ӯм „нҢҢ
    o.step()        # лӢӨмқҢ лӢЁкі„

    # print(f"y = {a.item()}x + {b.item()}")    # aлҠ” 10мңјлЎң bлҠ” 2лЎң к°Җкі мһҲлҠ”м§Җ нҷ•мқё
```
