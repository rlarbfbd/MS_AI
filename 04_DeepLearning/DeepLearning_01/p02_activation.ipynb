```python
# activation(활성화 함수)
#   Unit Step Function : 음수 들어오면 0, 양수 들어오면 1 -> 너무 빡빡함(값이 0과 1 밖에 없어 극단적), 0 부분이 잘려잇어서(불연속) 극한 값이 없음 -> 미분 불가 -> GD불가
#   Linear Activation : 들어오는거 그래도 나가게 = 아무 것도 안 한다
#   sigmoid: 1 / (1 + (e ** -x)) 
#       - USF를 부드럽게 펼친 느낌 -> USF의 단점들 해결(극단적이지 않음, 미분 가능)
#       - 하지만 미분해서 기울기가 나올 수 있는게 최대 1/4 -> 역전파가 진행되어서 1/4 * 1/4 * ... = 0에 가깝게 되어버림 -> 인공신경망의 뒷부분은 결과에 영향을 어느정도 주지만 앞부분은 영향을 거의 줄 수 없음 -> 몇 층 짜리 DNN을 만들어도 수학적으로 계산을 하면 결국 2층짜리 ANN이 나오게됨
#       => vanishing gradient 문제 발생
#           - 앞 층으로 갈수록 G 값이 거의 0
#           - 대세에 영향을 주지 못함
#   ReLU(Restricted Linear Unit)
#       음수 들어오면 0, 0도 0, 양수 들어오면 그대로
#       변형 ReLU 다수 존재
#   softmax: 활성화 함수라기 보다는 결론을 낼 때 씀
#       결과를 합이 1 되게 만들어줌(전기가 흘렀을 확률) 
#       ex) 결과 [70 5 25] -softmax-> [0.7 0.05 0.25]

# Non-linear activation이 필요한 이유(= 활성화 함수가 필요한 이유)
# - DNN으로 층 많이 만들면 짱
# 3층짜리 DNN을 만들어도 -> 2층짜리 ANN이랑 같은 상태
# 몇 층 짜리 DNN을 만들어도 수학적으로 계산을 하면 결국 2층짜리 ANN이 나오게됨
# => 그래서 활성화 함수가 필요하다
```
